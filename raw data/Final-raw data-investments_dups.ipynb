```python
temp = investments.drop(investments.columns.difference(['company_name', 'funded_at']), 1)
temp.set_index(['company_name'])
temp
```

```python
companyDateList = list(zip(temp['company_name'], temp['funded_at']))
```

```python
from collections import Counter

dups = dict(Counter(companyDateList))
dups
```

```python
def sec_rnd_energy_ix(row):
    company = row['company_name']
    date = row['funded_at']
    funding_amount = row['raised_amount_usd']
    search_key = tuple((company, date))
    div = dups[search_key]

    new_amount = funding_amount / div

    return float(new_amount)

investments['new_raised_amount_usd'] = investments.apply(sec_rnd_energy_ix, axis=1)
```
